```python
import cv2
import numpy as np
import time
from matplotlib.pyplot import imshow,title
%matplotlib inline
```

```python
image = cv2.imread('lena_gray.jpg')

# initialize sobel filter
sobelx = np.array([[-1, 0, 1], 
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype = np.float)
sobely = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]], dtype = np.float)

start_time = time.clock()

# applying gradient across X and Y direction
gx = cv2.filter2D(image, -1, sobelx)
gy = cv2.filter2D(image, -1, sobely)

# resultant image using :
# G = (Gx^2 + Gy^2)^0.5
g = np.sqrt(gx * gx + gy * gy)

time_diff = time.clock() - start_time

print('Result:\n2D convolution: execution time= {} '.format(time_diff))

g = g.astype('float32')
```

```python
imshow(image, cmap='gray');
title('Original Image');
```

```python
imshow(gx, cmap='gray');
title('Image after G_x gradient');
```

```python
imshow(gy, cmap='gray');
title('Image after G_y gradient');
```

```python
imshow(g, cmap='gray');
title('Image after Resultant gradient');
```
