```python
import cv2
import numpy as np
from matplotlib.pyplot import imshow,title
%matplotlib inline
```

```python
img1 = cv2.imread('stp2.jpg',0);
imshow(img1,cmap='gray');
title('Original Image: ');
```

```python
# Mean filter
M = np.ones((3,3), dtype=np.float32)/9
```

```python
H = cv2.filter2D(img1,-1,M)
imshow(H, cmap='gray');
title('Blurred Image: Image after applying Mean Kernal K = 1/9');
```
